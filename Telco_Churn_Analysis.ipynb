## Step 1: Load Dataset


```python

import pandas as pd
import numpy as np

df = pd.read_csv(
    r"C:\Users\architmehta\Telco_Churn_Project\Data\forSQL.csv"
)

print(df.head())
```

## Step 2: Explore Dataset Structure

```python
print("Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nInfo:")
print(df.info())
```

## Step 3: Check Missing Values

```python
print("Missing Values:")
print(df.isnull().sum())
```

## Step 4: Clean the Data

```python
# Remove completely empty rows
df = df.dropna(how='all')

# Remove rows with missing customerID
df = df[df['customerID'].notna()]

print(df.shape)
```

## Step 5: Convert Data Types

```python
df['TotalCharges'] = pd.to_numeric(
    df['TotalCharges'],
    errors='coerce'
)

print(df.isnull().sum())

# Remove rows with missing values
df = df.dropna()

print("Final Shape:")
print(df.shape)
```

## Step 6: Customer Segmentation

```python
df['CustomerSegment'] = np.where(
    df['tenure'] < 12,
    'New Customer',
    'Existing Customer'
)

print(df[['tenure', 'CustomerSegment']].head())

print(df['CustomerSegment'].value_counts())
```

## Step 7: Analyze Churn by Customer Segment

```python
segment_churn = pd.crosstab(
    df['CustomerSegment'],
    df['Churn'],
    normalize='index'
) * 100

print(segment_churn)
```

## Step 8: Analyze Churn by Contract Type

```python
contract_churn = pd.crosstab(
    df['Contract'],
    df['Churn'],
    normalize='index'
) * 100

print(contract_churn)
```

## Step 9: Analyze Churn by Payment Method


```python
payment_churn = pd.crosstab(
    df['PaymentMethod'],
    df['Churn'],
    normalize='index'
) * 100

print(payment_churn)
```

## Step 10: Analyze Churn by Internet Service

```python
internet_churn = pd.crosstab(
    df['InternetService'],
    df['Churn'],
    normalize='index'
) * 100

print(internet_churn)
```

## Step 11: Export Clean Dataset

```python
df.to_csv(
    r"C:\Users\architmehta\Telco_Churn_Project\outputs\cleaned_telco.csv",
    index=False
)

print("Cleaned dataset saved successfully!")
```

## Step 12: Business Insights

```python
print("Business Insights\n")

print(
    "Highest Churn Customer Segment:",
    segment_churn['Yes'].idxmax()
)

print(
    "Highest Churn Contract Type:",
    contract_churn['Yes'].idxmax()
)

print(
    "Highest Churn Payment Method:",
    payment_churn['Yes'].idxmax()
)

print(
    "Highest Churn Internet Service:",
    internet_churn['Yes'].idxmax()
)
```

## Step 13:Show Churn Percentages in Insights

```python
print(
    f"Customer Segment: {segment_churn['Yes'].idxmax()} "
    f"({segment_churn['Yes'].max():.2f}%)"
)

print(
    f"Contract Type: {contract_churn['Yes'].idxmax()} "
    f"({contract_churn['Yes'].max():.2f}%)"
)

print(
    f"Payment Method: {payment_churn['Yes'].idxmax()} "
    f"({payment_churn['Yes'].max():.2f}%)"
)

print(
    f"Internet Service: {internet_churn['Yes'].idxmax()} "
    f"({internet_churn['Yes'].max():.2f}%)"
)
```

# Business Conclusion:-
Based on the churn analysis, several key patterns were identified:

1. New Customers have the highest churn rate compared to existing customers. This indicates that customers are more likely to leave during their first year of service. Improving onboarding and early customer engagement could help reduce churn.


2. Month-to-Month Contract customers show the highest churn rate among all contract types. Customers with longer contracts (one-year and two-year) tend to be more loyal and less likely to leave.


3. Customers using Electronic Check as their payment method have the highest churn rate. This may indicate lower customer commitment or a less convenient payment experience compared to automatic payment methods.


4. Fiber Optic Internet customers have the highest churn rate among internet service categories. Although Fiber Optic provides faster internet speeds, customers may be dissatisfied with pricing, service quality, or customer support.


Recommendations:-

1. Focus retention efforts on new customers, especially during the first 12 months.


2. Encourage customers to switch from month-to-month contracts to longer-term contracts through discounts and loyalty programs.


3. Promote automatic payment methods such as bank transfer or credit card auto-pay.


4. Investigate the reasons behind the high churn rate among Fiber Optic customers and improve service quality and customer satisfaction.


Final Conclusion:-

1. Customer churn is primarily driven by:

2. Short customer tenure

3. Month-to-month contracts

4. Electronic check payment method

5. Fiber Optic internet service

By targeting these high-risk customer groups with retention strategies, the company can reduce churn, improve customer satisfaction, and increase long-term revenue.
